```python
import csv
import pandas as pd
import os
```

```python
df_comment = pd.read_csv('test.csv',sep = "\t", header = None, encoding = "utf8")
df_comment = df_comment.drop(0, axis = 1)
df_comment = df_comment.rename(index = int, columns = { 0 : "comment"})
df_comment = df_comment.rename(index = int, columns = { 1 : "result"})
display(df_comment)
```

```python
paths = [['aclImdb/train/neg/', -1], ['aclImdb/train/pos/', 1]]

result = {"comment": [], "score": []}

for path in paths:
    for filename in os.listdir(path[0]):
        data = open(path[0] + filename).read()
        data = data.replace("\n", "")
        result["comment"].append(data)
        result["score"].append(path[1])

data_df = pd.DataFrame(result) 
data_df.to_csv('train.csv', sep = "\t")
```
